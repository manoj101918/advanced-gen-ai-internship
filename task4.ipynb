Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False, "Recharge amount must be at least ₹50."

    if amount not in valid_plans:
        return False, f"Invalid plan. Available plans: {valid_plans}"

    return True, "Recharge successful!"


# Retry Logic
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        is_valid, message = validate_recharge(amount)

        print(message)

        if is_valid:
            break   # Exit loop if successful

    except ValueError:
        print("Please enter a valid numeric amount.")
```

Problem 2: Inventory Reorder Alert System

```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert  (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK  (Stock = {quantity})")


# Sample Dictionary
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 8,
    "Oil": 30,
    "Salt": 14
}

check_stock(inventory)
```

 Problem 3: Student Result Processing System

```python
def evaluate_student(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"

    return average, result


# Example Usage
marks_list = [60, 45, 70, 50, 40]

avg, status = evaluate_student(marks_list)

print("Average Marks:", avg)
print("Result:", status)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak:
        fare += fare * 0.25   # 25% extra

    return fare


# Retry System
while True:
    try:
        distance = float(input("Enter distance in km: "))

        peak_input = input("Is it peak hour? (yes/no): ").lower()
        peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, peak)

        print(f"Total Cab Fare: ₹{total_fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator ")
            break

    except ValueError:
        print("Please enter valid numeric distance.")
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    # Decide eligibility
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"


# Example Usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

percent, status = check_attendance(attendance)

print("Attendance Percentage:", percent)
print("Status:", status)
```

Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password  (Minimum length is 8)"

    has_digit = False
    has_special = False

    # Loop over string
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Check conditions
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


# Example Usage
pwd = input("Enter password: ")
result = check_password_strength(pwd)
print(result)
```
