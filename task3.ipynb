Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find employees with highest score
top_employees = []

for name, score in employees.items():
    if score == max_score:
        top_employees.append(name)

# Display result
print("Highest Performance Score:", max_score)
print("Employees Eligible for Bonus:", top_employees)
```

Problem Statement 2: Search Query Keyword Analysis


```python
sentence = "Buy mobile phone buy phone online"
sentence = sentence.lower()
words = sentence.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

#  Extract and count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

#  Calculate average
average = sum(sales) / len(sales)

# Calculate threshold
threshold = average * 1.3

#  Detect spike days
for i, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {i+1}: {sale}")
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

frequency = {}

# Count occurrences
for user in user_ids:
    frequency[user] = frequency.get(user, 0) + 1

#  Display duplicates
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
